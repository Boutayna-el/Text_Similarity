Import Libraries  

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
import string 
```

```python
import torch
import transformers 
from transformers import BertModel, BertTokenizer
```

Upload data 

```python
data1 = pd.read_csv('EmployeeSampleData.csv')
data1.head()
```

```python
data2 = pd.read_csv('EmployeeSampleDataMatched.csv')
data2.head()
```

Text Preprocessing 

```python
#Data Cleaning data1
# Remove duplicates
data1 = data1.drop_duplicates()
data1 = data1.dropna()
```

```python
# Drop unnecessary columns
data1 = data1[['EEID','Full Name','Hire Date','Annual Salary']]
```

```python
data1.head()
```

```python
#Data Cleaning data2
data2 = data2.drop_duplicates()
data2 = data2.dropna()
```

```python
# Drop unnecessary columns
data2 = data2[['EEID','Full Name','Hire Date','Annual Salary']]
```

```python
data2.head()
```

```python
# Lowercasing data1
for column in data1:
    data1[column] = data1[column].str.lower()
data1.head()
```

```python
# Lowercasing data2
for column in data2:
    data2[column] = data2[column].str.lower()
data2.head()
    
```

```python
#function to remove punctuation 
def remove_punctuation(text):
    # Convert the list element to a string
    text = " ".join(text)
    # Create a translation table to remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    # Use the translation table to remove punctuation from the text
    text_without_punct = text.translate(translator)
    return text_without_punct
#remove punctuation data2
for column in data2:
    data2[column]=data2[column].apply(remove_punctuation)
#remove punctuation data1
for column in data1:
    data1[column]=data1[column].apply(remove_punctuation)
    
```

```python
data1.head()
```

```python
data2.head()
```

Calcul similarity


```python
# Load pre-trained model and tokenizer
def get_bert_embeddings(texts):
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].numpy()  # Using the [CLS] token's output
```

```python
embeddings_data1 = {}
for column in data1:
    texts = data1[column].tolist()
    embeddings_data1[column] = get_bert_embeddings(texts)
```

```python
print("BERT embeddings for Data1:")
for column, embeddings in embeddings_data1.items():
    print(f"{column} embeddings:")
    print(embeddings)
```

```python
embeddings_data2 = {}
for column in data2:
    texts = data2[column].tolist()
    embeddings_data2[column] = get_bert_embeddings(texts)
```

```python
print("BERT embeddings for Data2:")
for column, embeddings in embeddings_data2.items():
    print(f"{column} embeddings:")
    print(embeddings)
```

Calcul cosine similarity

```python
columns_to_compare = ['EEID', 'Full Name', 'Hire Date','Annual Salary']
# Calculate cosine similarity between selected columns in data1 and corresponding columns in data2
for column in columns_to_compare:
    embeddings_data1_column = embeddings_data1[column]
    embeddings_data2_column = embeddings_data2[column]
    sim_score = cosine_similarity(embeddings_data1_column, embeddings_data2_column)
    print(f"Cosine similarity between '{column}' in data1 and data2:")
    print(sim_score)
```

thresholding and stockage the errors

```python
# List to store dissimilarity rows
dissimilarity_data = []
def classify_similarity(sim_score, threshold):
    return np.where(sim_score >= threshold, 1, 0)

threshold_value = 0.95
for column in columns_to_compare:
    embeddings_data1_column = embeddings_data1[column]
    embeddings_data2_column = embeddings_data2[column]
    sim_scores = cosine_similarity(embeddings_data1_column, embeddings_data2_column)

    # Classify similarity based on the threshold for the current column
    similarity_classes = classify_similarity(sim_scores, threshold_value)

    print(f"Cosine similarity between '{column}' in data1 and data2:")
    print(sim_scores)
    print(f"Binary similarity scores between '{column}' in data1 and data2:")
    print(similarity_classes)
    print(f"For the binary similarity scores '{column}'column:")
    for (i, j), element in np.ndenumerate(similarity_classes):
        if element == 0:
            print(f"Row {i + 1}, column {j + 1} are dissimilar.")
             # Append the dissimilarity row to the list
            dissimilarity_data.append([i + 1, j + 1, column])
            
```

```python
dissimilarity_df = pd.DataFrame(dissimilarity_data, columns=['Row', 'Column', 'Dissimilar_Column'])
# Print the entire DataFrame
print("\nDissimilarity DataFrame:")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(dissimilarity_df)
```
